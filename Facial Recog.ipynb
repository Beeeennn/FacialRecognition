```python
import face_recognition
import cv2
import numpy as np
from time import sleep as s
from playsound import playsound
length_time_gone = 500
bentimes =length_time_gone
joetimes =length_time_gone
ameliatimes =length_time_gone
nathantimes =length_time_gone
# This is a super simple (but slow) example of running face recognition on live video from your webcam.
# There's a second example that's a little more complicated but runs faster.

# PLEASE NOTE: This example requires OpenCV (the `cv2` library) to be installed only to read from your webcam.
# OpenCV is *not* required to use the face_recognition library. It's only required if you want to run this
# specific demo. If you have trouble installing it, try any of the other demos that don't require it instead.

# Get a reference to webcam #0 (the default one)
video_capture = cv2.VideoCapture(0)

# Load a sample picture and learn how to recognize it.
ben_image = face_recognition.load_image_file("Ben.jpg")
ben_face_encoding = face_recognition.face_encodings(ben_image)[0]

nathan_image = face_recognition.load_image_file("Nathan.jpg")
nathan_face_encoding = face_recognition.face_encodings(nathan_image)[0]

# Load a second sample picture and learn how to recognize it.
joe_image = face_recognition.load_image_file("Joe.jpg")
joe_face_encoding = face_recognition.face_encodings(joe_image)[0]

amelia_image = face_recognition.load_image_file("Amelia.jpg")
amelia_face_encoding = face_recognition.face_encodings(amelia_image)[0]

# Create arrays of known face encodings and their names
known_face_encodings = [
    ben_face_encoding,
    joe_face_encoding,
    amelia_face_encoding,
    nathan_face_encoding
    
]
known_face_names = [
    "Ben",
    "Joe",
    "Amelia",
    "Nathan"
]

while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()
    name = 'l'

    # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
    rgb_frame = frame[:, :, ::-1]

    # Find all the faces and face enqcodings in the frame of video
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    # Loop through each face in this frame of video
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # See if the face is a match for the known face(s)
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)

        name = "Unknown"
        # If a match was found in known_face_encodings, just use the first one.
        # if True in matches:
        #     first_match_index = matches.index(True)
        #     name = known_face_names[first_match_index]

        # Or instead, use the known face with the smallest distance to the new face
        face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
        best_match_index = np.argmin(face_distances)
        if matches[best_match_index]:
            name = known_face_names[best_match_index]
            if name == "Ben":
                if bentimes == length_time_gone:                 
                    playsound('Hello_ben.mp3')
                bentimes=0        
            else:
                bentimes = min(bentimes+1,length_time_gone)
                
                
            if name == "Nathan":
                if nathantimes == length_time_gone:                 
                    playsound('Hello_Nathan.mp3')
                nathantimes=0        
            else:
                nathantimes = min(nathantimes+1,length_time_gone)
                  
                
            if name=='Joe':
                
                if joetimes == length_time_gone:
                    playsound('Hello_daddy.mp3')
                joetimes = 0
            else:
                joetimes =min(joetimes+1,length_time_gone)
                
                
                
            if name=='Amelia':
                
                if ameliatimes == length_time_gone:
                    playsound('Amelia.mp3')
                ameliatimes = 0
            else:
                ameliatimes =min(ameliatimes+1,length_time_gone)
        else:
            name = 'no-one'
        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)
        print(name)
    # Display the resulting image
    cv2.imshow('Video', frame)
    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    bentimes = min(bentimes+1,length_time_gone)
    joetimes =min(joetimes+1,length_time_gone)
    ameliatimes =min(ameliatimes+1,length_time_gone)
    nathantimes =min(nathantimes+1,length_time_gone)
    print(bentimes)
# Release handle to the webcam
video_capture.release()
cv2.destroyAllWindows()
```
